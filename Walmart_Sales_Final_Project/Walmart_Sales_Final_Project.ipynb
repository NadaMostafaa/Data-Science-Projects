```python
WalmartData["Date"] = pd.to_datetime(WalmartData.Date, format="%d-%m-%Y")
WalmartData['Month'] = pd.to_datetime(WalmartData['Date'], format="%d-%m-%Y").dt.month


conditions=[
    (WalmartData['Holiday_Flag']==1) & (WalmartData['Month']==2),
    (WalmartData['Holiday_Flag']==1) & (WalmartData['Month']==9),
    (WalmartData['Holiday_Flag']==1) & (WalmartData['Month']==11),
    (WalmartData['Holiday_Flag']==1) & (WalmartData['Month']==12),
    (WalmartData['Holiday_Flag']==0)
]
choices=['Super Bowl','Labour Day','Thanksgiving','Christmas','Non Holidays']
holiday_analysis_df = WalmartData.drop(['Store', 'Month'], axis='columns')
holiday_analysis_df['Holiday']=np.select(conditions,choices)
mean_df = holiday_analysis_df.groupby('Holiday').agg(Mean_Weekly_Sales=('Weekly_Sales','mean')).round()
mean_df.plot(kind='bar', y='Mean_Weekly_Sales', figsize=(10,10))
mean_df.sort_values(by='Mean_Weekly_Sales',ascending = False)
```

```python
#Monthly sales
monthly = WalmartData.groupby(pd.Grouper(key='Date', freq='1M')).sum()# groupby each 1 month
monthly=monthly.reset_index()
fig, ax = plt.subplots(figsize=(10,8))
X = monthly['Date']
Y = monthly['Weekly_Sales']
plt.plot(X,Y)
plt.title('Month Wise Sales')
plt.xlabel('Monthly')
plt.ylabel('Weekly_Sales')
plt.show()
```

```python
#Semester Sales
Semester = WalmartData.groupby(pd.Grouper(key='Date', freq='6M')).sum()
Semester = Semester.reset_index()
fig, ax = plt.subplots(figsize=(10,8))
X = Semester['Date']
Y = Semester['Weekly_Sales']
plt.plot(X,Y)
plt.title('Semester Wise Sales')
plt.xlabel('Semester')
plt.ylabel('Weekly_Sales')
plt.show()
```

```python
WalmartData['Year'] = pd.to_datetime(WalmartData['Date'], format="%d-%m-%Y").dt.year
annually_df = WalmartData.groupby(['Year'], as_index=False).agg(Weekly_Sales=('Weekly_Sales','sum'))
sns.barplot(x = 'Year',y = 'Weekly_Sales',data = annually_df)
plt.title('Weekly Sales per Year')
plt.xlabel('Year')
plt.ylabel('Weekly Sales')
plt.show()
```

```python
WalmartData['month_year'] = pd.to_datetime(WalmartData['Date']).dt.to_period('M')

monthly_df = WalmartData.groupby(['month_year'], as_index=False).agg(Weekly_Sales=('Weekly_Sales','sum'))
monthly_df.head()
fig, ax = plt.subplots(figsize=(30,5))
sns.barplot(x = 'month_year',y = 'Weekly_Sales',data = monthly_df,ax=ax)
plt.title('Weekly Sales each month of the year')
plt.xlabel('Months')
plt.ylabel('Weekly Sales')
plt.show()
```
